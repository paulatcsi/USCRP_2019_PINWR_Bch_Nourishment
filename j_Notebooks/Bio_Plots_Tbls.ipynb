### Declare function _LoadBaseDataFrame_ to populate Pandas dataframe with fresh and use-specific, data for plots and/or analysis...

```python
def LoadBaseDataFrame(dp, f):
    df_ = pd.read_csv(dp+f, na_values=-9999)

    # data dates back to 2007 in this dataset but it's only consistent from 2012 onward...at least
    # for our purposes here. So, we'll first trim the data down from 2007 - 2020 to 2012 - 2020:
    df_ = df_.loc[df_['year'] >= 2012].copy()
    
    # add area (control vs treatment) column:
    df_.loc[df_['transect_ID'].str[0] == 'C', 'area'] = 'Control Beach'
    df_.loc[df_['transect_ID'].str[0] == 'T', 'area'] = 'Replenished Beach'

    # add number of transects to each survey for use as divisor in computing abundance densities (with a call
    # to the function ComputeSpeciesAbundanceDensity()
    df_.loc[(df_['originator'] == 'NCDOT') & (df_['area'] == 'Control Beach'), 'transect_cnt'] = 6
    df_.loc[(df_['originator'] == 'NCDOT') & (df_['area'] == 'Replenished Beach'), 'transect_cnt'] = 5
    df_.loc[(df_['originator'] == 'USFWS') | (df_['originator'] == 'CSI'), 'transect_cnt'] = 11
                                           
    # add Survey column
    month =df_['sample_date'].str[5:7]
    year = df_['sample_date'].str[0:4]
    df_['survey'] = year+'_'+month
    
    # add Survey key column  - this is an identifier that renders a set of records uniques to a specifc survey
    # as conduted by one of the three originators (NCDOT, USFWS, or CSI) along with its month and year of
    # occurance
    df_['survey_key'] = df_['originator']+'_'+year+'_'+month

    # drop treatment transects 12 - 16 (these are too much modified due to freq, storm damage):
    df_ = df_.loc[(df_['transect_ID'] != 'T12') & (df_['transect_ID'] != 'T13') & \
                  (df_['transect_ID'] != 'T14') & (df_['transect_ID'] != 'T15') & \
                  (df_['transect_ID'] != 'T16')].copy()
  
   # combine (sum) the three E. talpoida sizes classes into a single abundance for E. talpoida (I (Paul) elect
    # to do this because I feel I lack the requisite expertise in take on the task of explaining the why and the what
    # that this paritioning might reveal to the better prepared scientist...)
    
    df_['E_talpoida'] = df_[['sm_mole_crabs','md_mole_crabs','lg_mole_crabs']].sum(axis=1)
    
    # pool the replicate bio-counts for the USFWS and CSI collected data (NCDOTs data is already pooled)
    #   pool on the field sampleID -> domain:B1, B2, B3
    
    df = df_.groupby(['year', 'season', 'area']).agg({
        'originator':'first',
        'transect_ID':'first',
        'transect_cnt':'first',
        'survey':'first',
        'sample_date':'first',
        'ghost_crabs':'sum',
        'E_talpoida':'sum',
        'coquina_clams':'sum',
        'amphipods':'sum',
        'worms':'sum'
    }).reset_index()
    
    ### groupby spare parts:'survey_key','originator', 'transect_ID'         'area':'first', 'year':'first', 'season':'first',
    ###         'sample_ID':'sum',
    
    #df.drop('sample_ID', axis=1, inplace=True)
    
    # add season_lbl:
    df['survey_lbl'] = df['season']+' '+df['year'].astype('str')
    
    # add survey_datetime column where value is a Python date time object:
    df['Survey_Datetime'] = pd.to_datetime(df['sample_date'])
    
    # add Julian Date, and add 0.5 to bring date to whole day (Julian dates start counting at mid-day)
    df['DateJ'] = pd.DatetimeIndex(df['Survey_Datetime']).floor('d').to_julian_date()+0.5
    
    # rename a column or two in the dataframe before releasing it into the wild:
    df.rename(columns={'coquina_clams':'D_variabilis'}, inplace=True)
    df.rename(columns={'ghost_crabs':'O_quadrata'}, inplace=True)
    df.rename(columns={'worms':'S_squamata'}, inplace=True)
    df.rename(columns={'amphipods':'Amphipods'}, inplace=True)
    
    
    # finally, sort the df first on the survey date and then on beach area
    df.sort_values(['Survey_Datetime','area'], inplace=True)
    
    return(df)
```

### Function: ComputeSpeciesAbundanceDensity:

As the number of transects where counting occurred during the 8 years of data collection varied***, both in the Control and Impacted areas surveyed, it was necessary to adjust the raw counts so to normalize the data for comparison. This is the purpose of this function: to compute the abundance, as per:

$$ abundance = {n \over transect\_cnt} $$

where:
- n = total numbers of species counted per survey per area
- transect_cnt = number of transects included in the per survey area count

to ensure that results from the two different areas could be equitably (w/o bias) compared. 

*** NCDOT surveyed at 0.2 mile intervals while the USFWS and CSI sampled at 0.1 mile intervals. Moreover, storms occassionally removed parts of the beach in the study area making survey impossible in those locations. 

```python
def ComputeSpeciesAbundanceDensity(df):
    '''
  
    '''
    df['O_quadrata'] = df['O_quadrata'] / df['transect_cnt']
    df['E_talpoida'] = df['E_talpoida'] / df['transect_cnt']
    df['D_variabilis'] = df['D_variabilis'] / df['transect_cnt']
    df['S_squamata'] = df['S_squamata'] / df['transect_cnt']
    df['Amphipods'] = df['Amphipods'] / df['transect_cnt']
    return(df)
```

## PERMANOVAs

Some interesting questions to ask along the way:
- hat is the total abundance for each species in each area?
- what is the total abundance for each species for each area, partitioned for before and after replenishment?

For the ANOVA itself:
- is there a significant difference between abundances in the control and replenishment areas before replenishment?
- is there a significant difference between abundances in the control and relenishment areas after replenishment?
- is there a significant difference between abundances in control area before versus after replenishment?
- is there a significant difference between abundances in the replenished area before versus after replenishment?


Too, remember that you can't simply look at raw population counts because the number of transects surveyed did vary between surveys. For the data supplied by the USFWS and CSI, more often than not we have a equitable 11 transects impacted and another 11 in control, but from time to time the beach was sufficiently damaged or in places impassible, both of which prevented counts along one or more of those 22 transects. More oftne than not this was a problem in the relenished area, but not always. Furthermore, the NCDOT data is based on transect spacing of 0.2 miles (USFWS and CSI used a 0.1 mile spacing) which means that most of the DOT data is based not on 22 transects but rather 10. The fix is to compute the relative abundance density (avg. count per transect) in place of raw counts. The function ComputeSpeciesAbundance(df, 'Annelids') is designed to do this computation for us--but it works on a species by species basis...
