```python
data_path = '../Final_Data_Sets/storms.csv'

df = pd.read_csv(data_path)
df
```

```python

g = sns.catplot(x='Year', hue='Type', data=df,
                 kind='count',
                height=7, aspect=2.7,
                legend=False)

g.set_axis_labels('', 'Number of Storms', size=18)
g.despine(left=True)

g.set_xticklabels(rotation=90, fontsize=18)
g.set(yticks=[0,1,2,3,4,5])
g.set_yticklabels(['0','1','2','3','4','5'], fontsize=15)

plt.title('Hurriances and Tropical Storms that passed within 300 miles of the Pea Island Study Area', fontsize=21)
plt.legend(loc='center', fontsize=12)
plt.tight_layout()

plt.savefig('../Final_Plots/Storms_2000_2020.png')
;
```
