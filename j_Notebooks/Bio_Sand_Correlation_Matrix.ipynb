# Let's do turtles

```python
df_turts = pd.read_csv('../Final_Data_Sets/Pea Island Sea Turtle Summary 1980 2020 AOI.csv' )
df_turts = df_turts.loc[df_turts['Year'] >= 2012].copy()

df_turts.loc[df_turts['Year'] < 2014, 'Epoch'] = '2012-2013'
df_turts.loc[(df_turts['Year'] >= 2014) & (df_turts['Year'] <= 2020), 'Epoch'] = '2014-2020'

df_turts.drop(['Beach','County','Nest #','Latitude','Longitude','Location','Relocati_3','Relocati_4','Relocati_5'], axis=1, inplace=True)

df_turtles = df_turts.groupby(['Year','Area']).agg({
    'Species':'first',
    'Relocation':'first',
    'Clutch Cou':'mean',
    'Hatch Succ':'mean',
    'Emergence':'mean',
    'Epoch':'first'
}).reset_index()

df_turtles #['Activity D'].unique()
```

```python
df_sands = df_sed.groupby(['year','season','area']).agg({
    '%gravels_S4':'mean',
    '%c_sands_S4':'mean',
    '%m_sands_S4':'mean',
    '%f_sands_S4':'mean',
    'epoch':'first'
}).reset_index()

# extract ust summer season recs:
df_sands = df_sands.loc[df_sands['season'] == 'summer']

# rename areas designated as Impacted to Replenished:
df_sands.loc[df_sands['area'] == 'Impacted', 'area'] = 'Replenished'
df_sands
```

### Correlation Table:

```python
pd.concat([df_turtles,df_sands],axis=1).corr()
```

### Correlation Heat Map:

```python
plt.figure(figsize=(24,24))

msk = np.triu(np.ones_like(pd.concat([df_turtles,df_sands],axis=1).corr(), dtype=bool))
sns.heatmap(pd.concat([df_turtles,df_sands],axis=1).corr(), annot=True, mask=msk, cbar=False, cmap='BrBG')

plt.savefig(
```
